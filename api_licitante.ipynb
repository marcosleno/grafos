### 2. Obtendo  a base de dados dos licitantes de 185 municipios (2010 a 2023).

```python
# https://api-dados-abertos.tce.ce.gov.br/licitacoes?codigo_municipio=057&data_realizacao_autuacao_licitacao=2010-01-01_2010-12-31
# url = "https://api-dados-abertos.tce.ce.gov.br/municipios"
# https://api-dados-abertos.tce.ce.gov.br/licitantes?codigo_municipio=057&data_realizacao_licitacao=2010-01-01_2010-12-31&quantidade=100&deslocamento=0
url1 = "https://api-dados-abertos.tce.ce.gov.br/licitantes?codigo_municipio="
url2 = "&data_realizacao_licitacao="
url3 = "&quantidade=100&deslocamento=0"
headers = {
    "accept": "application/json"
}
```

```python
for i in codigo_municipios:
    for j in data_licitacao:
        sucesso = False
        retries = 0
        while not sucesso:
            response = requests.get(url1 + i + url2 + j + url3, headers=headers)
            if response.status_code == 200: # verifica se o servidor respondeu
                sucesso = True
                data = response.json()
                if len(data['data']) > 0:
                    with open(i + "_" + j + "_" + "licitante.json", "w") as json_file:
                        json.dump(data, json_file, indent=4)
            else:
                print("Erro " + i + "_" + j)
                time.sleep(10)
                retries += 1
                if retries >= 2: # realiza at√© 2 tentativas, em intervalos de 10 segundos
                    sucesso = True
```
