```python
import networkx as nx
import matplotlib.pyplot as plt
import json
from networkx.algorithms import bipartite
```

```python
nome_arquivo = "002_2010-01-01_2010-12-31_licitante.json"
   
with open(nome_arquivo, 'r') as arquivo:
    dados = json.load(arquivo)

lista_links = [(item["numero_licitacao"], item["nome_negociante"]) for item in dados["data"]]
```

```python
G = nx.Graph()

licitacoes = [(item["numero_licitacao"]) for item in dados["data"]]
licitantes = [(item["nome_negociante"]) for item in dados["data"]]

G.add_nodes_from(licitacoes, bipartite=0)
G.add_nodes_from(licitantes, bipartite=1)

lista_links = [(item["numero_licitacao"], item["nome_negociante"]) for item in dados["data"]]

G.add_edges_from(lista_links)
```

```python
edges = G.edges()
nx.draw_networkx(G, pos = nx.drawing.layout.bipartite_layout(G, licitacoes), width = 1, with_labels=True, font_size=8)
```
